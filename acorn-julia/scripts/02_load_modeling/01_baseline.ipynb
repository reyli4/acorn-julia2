```python
"""
Here we extract hourly zonal temperature data, which is needed
for the load model. We are working with the historical and future
TGW data, which is stored at `pu.tgw_path`. The hourly zonal temperatures
will be stored at `pu.project_path/data/climate/tgw`.
"""

# Required TGW variables
tgw_vars = ["T2C"]
tgw_var_str = tgw_vars[0]

# All TGW experiments
tgw_experiments = os.listdir(pu.tgw_path)

# Loop through all
for tgw_experiment in tgw_experiments:
    if os.path.isfile(
        f"{pu.project_path}/data/climate/tgw/zonal_{tgw_var_str}_{tgw_experiment}.csv"
    ):
        print(f"{tgw_experiment} already done")
        continue

    # Extract years
    tgw_experiment_yrs_str = tgw_experiment.split("_")[1:]
    tgw_years = np.arange(
        int(tgw_experiment_yrs_str[0]), int(tgw_experiment_yrs_str[1]) + 1
    )

    # Get paths of all hourly netcdf files
    tgw_paths = [
        glob(f"{pu.tgw_path}/{tgw_experiment}/hourly/*_{year}-*") for year in tgw_years
    ]
    tgw_paths = [path for sub_path in tgw_paths for path in sub_path]  # flatten

    # Parallelize with dask
    delayed = []

    for path in tgw_paths:
        delayed.append(dask.delayed(cu.tgw_to_zones)(path, tgw_vars))

    # Store
    df_out = pd.concat(dask.compute(*delayed))
    df_out.to_csv(
        f"{pu.project_path}/data/climate/tgw/zonal_{tgw_var_str}_{tgw_experiment}.csv"
    )
    print(tgw_experiment)
```

### Fit the multi-zone model

```python
# Create or load the multi-zone neural network model. This will train the model if it
# is not already stored at `modelpath`.
if not os.path.isfile(modelpath):
    # Specifiy model
    model = MLPRegressor(
        hidden_layer_sizes=(100, 100), early_stopping=True, max_iter=5000
    )

    # Initiate
    predictor = plm.MultiZoneLoadPredictor(model=model)

        # ---------- INSERT THIS NORMALIZATION ----------
    def _normalize_for_trainer(temp_data, load_data):
        td = temp_data.copy()
        ld = load_data.copy()

        # drop artifact cols, rename to expected names
        td = td.drop(columns=[c for c in ("level_0","index") if c in td.columns], errors="ignore")
        ld = ld.drop(columns=[c for c in ("level_0","index") if c in ld.columns], errors="ignore")
        td = td.rename(columns={"ZONE":"zone", "datetime":"time"})
        ld = ld.rename(columns={"ZONE":"zone", "datetime":"time"})

        # types
        td.loc[:, "time"] = pd.to_datetime(td["time"], utc=True, errors="coerce")
        ld.loc[:, "time"] = pd.to_datetime(ld["time"], utc=True, errors="coerce")
        td.loc[:, "zone"] = td["zone"].astype(str)
        ld.loc[:, "zone"] = ld["zone"].astype(str)

        # keep common zones & overlapping time window
        common = sorted(set(td["zone"]).intersection(ld["zone"]))
        td = td[td["zone"].isin(common)].copy()
        ld = ld[ld["zone"].isin(common)].copy()
        start = max(td["time"].min(), ld["time"].min())
        end   = min(td["time"].max(), ld["time"].max())
        td = td[(td["time"] >= start) & (td["time"] <= end)].copy()
        ld = ld[(ld["time"] >= start) & (ld["time"] <= end)].copy()

        # require full load coverage across zones per timestamp
        nz = len(common)
        good_times = ld.groupby("time")["zone"].nunique()
        good_times = good_times[good_times == nz].index
        td = td[td["time"].isin(good_times)].copy()
        ld = ld[ld["time"].isin(good_times)].copy()

        # lightly fill temp gaps (loads are not imputed)
        td_wide = (td.pivot_table(index="time", columns="zone", values="T2C")
                     .sort_index().interpolate(limit_direction="both"))
        td = td_wide.stack().rename("T2C").reset_index()
        return td, ld

    td, ld = _normalize_for_trainer(temp_data, load_data)
    # ---------- END INSERT ----------
    

    
    # Train
    metrics = predictor.train(td, ld, test_split=[2016], temp_varname="T2C")

    # Save model
    from pathlib import Path
    Path(modelpath).parent.mkdir(parents=True, exist_ok=True)
    predictor.save_model(modelpath)
else:
    # Load the model
    from python.utils import project_path
    import os
    modeldir  = f"{project_path}/data/load/baseline/models"
    modelpath = f"{modeldir}/load_model_tgw_multizone_NN.pkl"
    predictor = plm.MultiZoneLoadPredictor(model=None)
    predictor.load_model(modelpath)
    # Get the fit metrics
    metrics = predictor.results["metrics"]
```

## Generating new timeseries

```python
# Load the model
predictor = plm.MultiZoneLoadPredictor(model=None)
predictor.load_model(modelpath)
```

```python
# New climate input data
tgw_var_str = "T2C"
tgw_experiments = np.sort(os.listdir(tgw_path))

for tgw_experiment in tgw_experiments[:2]:
    store_path = f"{project_path}/data/load/baseline/simulated/{modelname}_{tgw_experiment}.parquet"
    if os.path.exists(store_path):
        continue

    # 1) raw TGW zonal file
    temp_data_path = f"{project_path}/data/climate/tgw/zonal_{tgw_var_str}_{tgw_experiment}.csv"

    # 2) normalize columns to exactly: zone, time, T2C
    df = pd.read_csv(temp_data_path)
    rename_map = {}
    if "ZONE" in df.columns: rename_map["ZONE"] = "zone"
    if "zone_name" in df.columns and "zone" not in df.columns: rename_map["zone_name"] = "zone"
    if "datetime" in df.columns and "time" not in df.columns: rename_map["datetime"] = "time"
    df = df.rename(columns=rename_map)

    # keep only needed cols in the right names/order
    df = df[["zone", "time", tgw_var_str]]

    # 3) write a normalized CSV and use THAT path
    base, _ = os.path.splitext(temp_data_path)
    norm_path = base + "__norm.csv"
    df.to_csv(norm_path, index=False)

    # 4) get bus-level load (NOTE: pass norm_path + temp_varname)
    df_bus_load = plm.baseline_load_to_bus(predictor, norm_path, temp_varname=tgw_var_str)

    # 5) save
    os.makedirs(os.path.dirname(store_path), exist_ok=True)
    df_bus_load.to_parquet(store_path, index=False)
```
